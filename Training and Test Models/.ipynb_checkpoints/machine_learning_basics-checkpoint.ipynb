```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv')
train.head()
print(train.shape[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

saleprice = pd.DataFrame(train['SalePrice']) #Response
gr = pd.DataFrame(train['GrLivArea']) #Predictor
#print(saleprice.shape)
#print(gr.shape)

#Split the data set into 80:20 train:test
gr_train, gr_test, saleprice_train, saleprice_test = train_test_split(gr, saleprice, test_size = 0.2)

##Train set
#saleprice_train = saleprice[:1168]
#gr_train = gr[:1168]

##Test set
#saleprice_test = saleprice[-292:]
#gr_test = gr[-292:]

print(saleprice_train.shape)
print(gr_train.shape)
print(saleprice_test.shape)
print(gr_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
#Import linear regression object
linreg = LinearRegression()

#Train the linear regression model
linreg.fit(gr_train, saleprice_train)

```

## Now that model is trained on data, it has estimated the coefficient and intercept based on the linear model: Saleprice = a x GrLivArea + b

## 1(a) Create appropriate datasets for Train and Test in an 80:20 ratio and fit a Linear Regression model on the Train set to predict SalePrice using GrLivArea. Print the coefficients of your model and plot the regression line.

**coefficient:** `104.27924399` 
**intercept:** `[22473.52894485]`



```python
#Find coefficient and intercept

print('coefficient:', linreg.coef_)
print('intercept:', linreg.intercept_)
```

```python
#Predict the relationship as a regression line
predicted_train_gr = gr_train
predicted_train_saleprice = linreg.coef_ * gr_train + linreg.intercept_

#Plot the line on the scatter plot
f, axes = plt.subplots(1,1, figsize = (16,8))
plt.scatter(gr_train, saleprice_train)
plt.plot(predicted_train_gr, predicted_train_saleprice, color='red', linewidth=3)
plt.show()
```


## 1(b) Check the Goodness of Fit of the model on the Train set and Prediction Accuracy of the model on the Test set. Print the metrics for Goodness of Fit and Prediction Accuracy that you think are appropriate in each scenario.

#### Explained Variance (R^2) = `0.4863017814411351`
#### Mean Squared Error (MSE)= `3256597129.866755`
#### Root Mean Squared Error (RMSE)= `57066.60257862522`

#### The goodness of fit (on the train set) is decent but not perfect or overfitted (R^2 =1), as it has an explained variance of 0.48

```python
# Check how good the predictions are on the (train data set)
# Metrics used: Explained Variance (R^2) and Mean Squared Error (MSE)

#Explained Variance
print('Explained Variance (R^2) =', linreg.score(gr_train, saleprice_train))

#Mean Squared Error
def MSE(actual, predicted):
    MSE = np.mean(np.square(np.array(actual) - np.array(predicted)))
    return MSE

MSE = MSE(saleprice_train, predicted_train_saleprice)
print('Mean Squared Error (MSE)=', MSE)


#Root Mean Squared Error
print('Root Mean Squared Error (RMSE)=', np.sqrt(MSE))

```

## 1(b) Check the Goodness of Fit of the model on the Train set and Prediction Accuracy of the model on the Test set. Print the metrics for Goodness of Fit and Prediction Accuracy that you think are appropriate in each scenario.

#### Explained Variance (R^2) = `0.565128206448618`
#### Mean Squared Error (MSE)= `2684909287.020331`
#### Root Mean Squared Error (RMSE)= `51816.110303845955`

#### The Prediction Accuracy (on the train set) is decent but not perfect or overfitted (R^2 =1), as it has an explained variance of 0.56
#### Since the RMSE (Root mean square error) is lower on the test model than on the training model, its prediction accuracy on saleprice is higher than that of the training data set.

```python
#Compute the predicted sale price based on test data set
predicted_test_saleprice = linreg.predict(gr_test)

#plot the scatter plot and linear regression line (test data)
f, axes = plt.subplots(1,1, figsize = (16,8))
plt.scatter(gr_test, saleprice_test, color='green')
plt.plot(gr_test, predicted_test_saleprice, color='red')
plt.show()
```

```python
# Check how good the predictions are on the (test data set)
# Metrics used: Explained Variance (R^2) and Mean Squared Error (MSE)

#Explained Variance
print('Explained Variance (R^2) =', linreg.score(gr_test, saleprice_test))

#Mean Squared Error
def MSE(actual, predicted):
    MSE = np.mean(np.square(np.array(actual) - np.array(predicted)))
    return MSE

MSE = MSE(saleprice_test, predicted_test_saleprice)
print('Mean Squared Error (MSE)=', MSE)


#Root Mean Squared Error
print('Root Mean Squared Error (RMSE)=', np.sqrt(MSE))
```

```python
saleprice = pd.DataFrame(train['SalePrice']) #Response
bsmt = pd.DataFrame(train['TotalBsmtSF']) #Predictor

#Split the data set into 80:20 train:test
bsmt_train, bsmt_test, saleprice_train, saleprice_test = train_test_split(bsmt, saleprice, test_size = 0.2)

#Import linear regression object
linreg = LinearRegression()

#Train the linear regression model
linreg.fit(bsmt_train, saleprice_train)

#Find coefficient and intercept

print('coefficient:', linreg.coef_)
print('intercept:', linreg.intercept_)


#Predict the relationship as a regression line (saleprice vs totalBsmt)
predicted_train_bsmt = bsmt_train
predicted_train_saleprice = linreg.coef_ * bsmt_train + linreg.intercept_

#Plot the line on the scatter plot
f, axes = plt.subplots(1,1, figsize = (16,8))
plt.scatter(bsmt_train, saleprice_train)
plt.plot(predicted_train_bsmt, predicted_train_saleprice, color='red', linewidth=3)
plt.show()


# Check how good the predictions are on the (train data set)
# Metrics used: Explained Variance (R^2) and Mean Squared Error (MSE)

#Explained Variance
print('Explained Variance (R^2) =', linreg.score(bsmt_train, saleprice_train))

#Mean Squared Error
def MSE(actual, predicted):
    MSE = np.mean(np.square(np.array(actual) - np.array(predicted)))
    return MSE

MSE = MSE(saleprice_train, predicted_train_saleprice)
print('Mean Squared Error (MSE)=', MSE)


#Root Mean Squared Error
print('Root Mean Squared Error (RMSE)=', np.sqrt(MSE))


#_________________________________________________________________________________________________


#Compute the predicted sale price based on test data set
predicted_test_saleprice = linreg.predict(bsmt_test)

#plot the scatter plot and linear regression line (test data)
f, axes = plt.subplots(1,1, figsize = (16,8))
plt.scatter(bsmt_test, saleprice_test, color='green')
plt.plot(bsmt_test, predicted_test_saleprice, color='red')
plt.show()


# Check how good the predictions are on the (test data set)
# Metrics used: Explained Variance (R^2) and Mean Squared Error (MSE)

#Explained Variance
print('Explained Variance (R^2) =', linreg.score(bsmt_test, saleprice_test))

#Mean Squared Error
def MSE(actual, predicted):
    MSE = np.mean(np.square(np.array(actual) - np.array(predicted)))
    return MSE

MSE = MSE(saleprice_test, predicted_test_saleprice)
print('Mean Squared Error (MSE)=', MSE)


#Root Mean Squared Error
print('Root Mean Squared Error (RMSE)=', np.sqrt(MSE))
```

```python
def upper_outliers(data):
    Q3 = np.percentile(data, 75)
    Q1 = np.percentile(data, 25)
    IQR = Q3 - Q1
    topwhisker = Q3 + 1.5*IQR
    return topwhisker

def lower_outliers(data):
    Q3 = np.percentile(data, 75)
    Q1 = np.percentile(data, 25)
    IQR = Q3 - Q1
    bottomwhisker = Q1 - 1.5*IQR
    return bottomwhisker

#Drop outlier datapoints of grlivarea
gr_upper = upper_outliers(gr)
gr_lower = lower_outliers(gr)
train = train[train.GrLivArea >= gr_lower]
train = train[train.GrLivArea <= gr_upper]

#Drop outlier datapoints of saleprice
saleprice_upper = upper_outliers(saleprice)
saleprice_lower = lower_outliers(saleprice)
train = train[train.SalePrice >= saleprice_lower]
train = train[train.SalePrice <= saleprice_upper]

# Re-create dataframes for grlivarea and saleprice
gr = pd.DataFrame(train['GrLivArea'])
saleprice = pd.DataFrame(train['SalePrice'])

```
